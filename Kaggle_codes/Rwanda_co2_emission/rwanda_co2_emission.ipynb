```python
import numpy as np
import pandas as pd


train = pd.read_csv(r'C:\Users\Reet\Downloads\dataset\Time_series\Kaggle\co2_emission\train.csv\train.csv')
test = pd.read_csv(r'C:\Users\Reet\Downloads\dataset\Time_series\Kaggle\co2_emission\test.csv\test.csv')

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
train.head()
```

```python

#train.columns[68]

for i in range(1,68):
    train[train.columns[i]].fillna(value=train[train.columns[i]].mean() ,inplace=True)
    
for i in range(1,68):
    test[test.columns[i]].fillna(value=test[test.columns[i]].mean() ,inplace=True)



```

```python
from sklearn.model_selection import train_test_split 
from sklearn import metrics
import matplotlib.pyplot as plt

x1 = train
x1 = x1.drop(['emission'], axis=1)
y1 = train[['emission']]
X_TEST = test


X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.15, random_state=1)
```

```python

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
#scaler = StandardScaler()
scaler = MinMaxScaler()

X_train_ID = X_train[X_train.columns[X_train.columns.isin(['ID_LAT_LON_YEAR_WEEK'])]]
X_train_ID_f = X_train[X_train.columns[~X_train.columns.isin(['ID_LAT_LON_YEAR_WEEK'])]]
X_train_scaled = scaler.fit_transform(X_train_ID_f)

X_test_ID = X_test[X_test.columns[X_test.columns.isin(['ID_LAT_LON_YEAR_WEEK'])]]
X_test_ID_f = X_test[X_test.columns[~X_test.columns.isin(['ID_LAT_LON_YEAR_WEEK'])]]
X_test_scaled = scaler.fit_transform(X_test_ID_f)

X_TEST_ID = X_TEST[X_TEST.columns[X_TEST.columns.isin(['ID_LAT_LON_YEAR_WEEK'])]]
X_TEST_ID_f = X_TEST[X_TEST.columns[~X_TEST.columns.isin(['ID_LAT_LON_YEAR_WEEK'])]]
X_TEST_scaled = scaler.fit_transform(X_TEST_ID_f)
```

```python
y_train = y_train.values
```

```python
y_train = scaler.fit_transform(y_train)
y_test = scaler.fit_transform(y_test)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X_train_scaled, y_train)

y_pred = regr.predict(X_test_scaled)
print("Prediction for test set: {}".format(y_pred))
print("Intercept: ", regr.intercept_)
print("Coefficients:")
list(zip(X_train_scaled, regr.coef_))
```

```python
### Actual pred

Y_PRED = regr.predict(X_TEST_scaled)
```
